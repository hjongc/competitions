```python
from sklearn.cluster import KMeans

cluster = KMeans(init="k-means++", n_clusters=8, n_init=10)
cluster.fit(df[X])
```

```python
df['cluster'] = cluster.predict(df[X])
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

inertias = []

for k in range(1,21):
    cluster = KMeans(n_clusters=k, n_init=10)
    cluster.fit(df[X])
    inertias.append(cluster.inertia_)
    print(f'n_cluster: {k}, inertia: {cluster.inertia_}')
    

plt.figure(figsize = (15,6))
plt.plot(range(1,21), inertias, '-o')
plt.xlabel('num of cluster(K)')
plt.ylabel('inertia')
plt.xticks(range(1,21))
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline


pca = PCA(n_components= 0.95)
trans_X = pca.fit_transform(df[X])
inertias = []
print(f'pca n_componets: {i}')

for k in range(1,21):
    cluster = KMeans(n_clusters=k, n_init=10)
    cluster.fit(trans_X)
    inertias.append(cluster.inertia_)

    
plt.figure(figsize = (15,6))
plt.plot(range(1,21), inertias, '-o')
plt.xlabel('num of cluster(K)')
plt.ylabel('inertia')
plt.xticks(range(1,21))
plt.show()
```
