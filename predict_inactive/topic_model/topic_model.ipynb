```python
import pandas as pd
```

```python
df = pd.read_excel('../dataset/2019_투입산출표_생산자가격_통합소분류.xlsx', skiprows=5)
```

```python
df = df.iloc[:-7, :-17]
```

```python
df
```

```python
df = df.rename(columns = {'Unnamed: 0': '코드', 'Unnamed: 1': '상품', '총공급계\n':'총공급계'})
```

```python
df
```

```python
df_out = df.set_index('상품')
df_out = df_out.iloc[:,1:]

df_out = df_out.apply(lambda x: (x[:]-min(x))/(max(x)-min(x)), axis=1)
df_out = df_out.add_prefix("To_")
df_out
```

```python
df_out = df_out.fillna(0)
```

```python
df_in = df.set_index('상품')
df_in = df_in.iloc[:,1:]
df_in = df_in.T


df_in = df_in.apply(lambda x: (x[:]-min(x))/(max(x)-min(x)), axis=1)
df_in = df_in.add_prefix("From_")
df_in
```

```python
df_out
```

```python
doc = pd.concat([df_in, df_out],axis=1)
doc
```

```python
import pandas as pd
import urllib.request
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
from sklearn.decomposition import TruncatedSVD

n_topics = 11
lsa_model = TruncatedSVD(n_components = n_topics)
lsa_model.fit_transform(doc)
```

```python
print(lsa_model.components_.shape)
```

```python
terms = doc.columns # 단어 집합. 5,000개의 단어가 저장됨.

# 각 주제에서 n개씩 단어 출력
# 토픽 모델링
def get_topics(components, feature_names, n=11):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(lsa_model.components_, terms)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# 토픽 개수 10개
lda_model = LatentDirichletAllocation(n_components=11, learning_method='online', random_state=777, max_iter=1)
lda_model.fit_transform(doc)
```
